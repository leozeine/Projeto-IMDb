```python
#urls
movies_list_kr = ['https://www.imdb.com/search/title/?country_of_origin=BR']
start = 'href=\'/'
for i in range(25000):
    try:
        response = requests.get(movies_list_kr[i])
        soup = BeautifulSoup(response.content, 'html.parser')
        nextl = str(soup.find_all('a', class_='lister-page-next next-page')[0])
        link = 'https://www.imdb.com/'+nextl[nextl.find('href=\"')+len(start):nextl.rfind('\">Next')]
        movies_list_kr.append(link)
    except:
        break
```

```python
movies_list_fr = ['https://www.imdb.com/search/title/?country_of_origin=FR']
start = 'href=\'/'
for i in range(110000):
    try:
        response = requests.get(movies_list_fr[i])
        soup = BeautifulSoup(response.content, 'html.parser')
        nextl = str(soup.find_all('a', class_='lister-page-next next-page')[0])
        link = 'https://www.imdb.com/'+nextl[nextl.find('href=\"')+len(start):nextl.rfind('\">Next')]
        movies_list_fr.append(link)
    except:
        break

```

```python
movies_names, movies_years, genres, links, rates, imdb_rates=[], [], [], [], [], []
flag = 0
urls_t=urls_fr[1975:]
for movies in urls_t:
    response = requests.get(movies)
    soup = BeautifulSoup(response.content, 'html.parser')
    movie_name = soup.find_all('h3', class_='lister-item-header')
    movie_year = soup.find_all('span', class_="lister-item-year text-muted unbold")
    for i in movie_name:
        box1 = i.get_text()
        box2 = i.get_text()
        box1 = re.findall('\\n.+\\n(.+)\\n.+\\n', box1)
        
        box2 = re.findall(r'\((\d.*?)\)', box2)

        if (len(box1)>=1 and len(box2)>=1):
            movies_names.append(box1[0])
        elif (len(box1)==0 and len(box2)>=1):
            movies_names.append(' ')
        elif (len(box1)>=1 and len(box2)==0):
            movies_names.append(box1[0])
        elif (len(box1)==0 and len(box2)==0):
            movies_names.append(' ')
        if (len(box2)>0):
            movies_years.append(box2[0])
        else:
            movies_years.append('')
   
    for i in range(len(soup.findAll(class_=['lister-item-header', 'genre']))):
        try:
            genre = re.findall(r'"genre">\n(.*?)<', soup.findAll(class_=['lister-item-header', 'genre'])[i].__str__())[0].split()
            for j in range(flag-1): 
                genres.append(['None'])
            genres.append(genre)
            flag=0
        except:
            flag+=1

    for i in range(len(soup.findAll(class_=['lister-item-header', 'ratings-bar']))):
        text = soup.findAll(class_=['lister-item-header', 'ratings-bar'])[i].get_text().split()
        if (i==len(soup.findAll(class_=['lister-item-header', 'ratings-bar']))-1 and text.count('Rate')==0 and text.count('this')==0):
          rates.append('None')
          imdb_rates.append('None')
        try:
            rating = text[text.index('Metascore')-1]
            rates.append(rating)
            imdb_rates.append(text[text.index('Rate')-1])
        except:
            if (text.count('Rate')==1 and text.count('this')==1):
                rates.append('None')
                imdb_rates.append(text[text.index('Rate')-1])
            elif (text.count('Rate')==0 and text.count('this')==0 \
                  and i!=0 \
                  and soup.findAll(class_=['lister-item-header', 'ratings-bar'])[i-1].get_text().split().count('Rate')==0):
                rates.append('None')
                imdb_rates.append('None')
        
        
    if (len(movies_names)%1000==0):
      print('movies ', len(movies_names))
      print('genres ', len(genres))
      print('rates ', len(rates))
      print('imdb_rates', len(imdb_rates))
    
    try:
      df = pd.DataFrame(data={'title': movies_names, 'genres':genres, 'year':movies_years, 'metacritic':rates, 'imdb_rating':imdb_rates})
    except:
      print(movies)
    if (len(movies_names)==30000):
      break
      
```

```python
df = pd.DataFrame(data={'year': movies_years, 'title':movies_names, 'genre':genres, 'metascore':rates, 'imdb_rating':imdb_rates})
```
