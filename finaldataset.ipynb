```python
def clean(genre):
    g = []
    for i in genre[genre.index('[')+1:genre.index(']')].split():
        '''if (i==genre[genre.index('[')+1:genre.index(']')].split()[-1]):
            g.append(i[:i.index('\'')])'''
        try:
            g.append(i[i.index('\'')+1:i.index(',')])
        except:
            g.append(i[i.index('\'')+1:-1])
    return g
```
